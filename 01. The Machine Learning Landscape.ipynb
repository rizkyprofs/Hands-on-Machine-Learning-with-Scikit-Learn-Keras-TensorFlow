## Apa Itu Machine Learning?

Machine Learning adalah ilmu (dan seni) memprogram komputer sehingga mereka dapat belajar dari data tanpa diprogram secara eksplisit.

*   **Definisi Arthur Samuel (1959)**: "Machine Learning adalah bidang studi yang memberi komputer kemampuan untuk belajar tanpa diprogram secara eksplisit."
*   **Definisi Tom Mitchell (1997)**: "Sebuah program komputer dikatakan belajar dari pengalaman E sehubungan dengan tugas T dan ukuran kinerja P, jika kinerjanya pada T, sebagaimana diukur oleh P, meningkat dengan pengalaman E."

**Contoh**: Filter spam adalah program ML.
*   **Tugas (T)**: Menandai spam untuk email baru.
*   **Pengalaman (E)**: Data pelatihan (contoh email spam dan non-spam).
*   **Ukuran Kinerja (P)**: Rasio email yang diklasifikasikan dengan benar (akurasi).

Mengunduh data, seperti Wikipedia, tidak serta merta membuat komputer belajar kecuali komputer tersebut menggunakan data tersebut untuk meningkatkan kinerja pada tugas tertentu.

## Mengapa Menggunakan Machine Learning?

Machine Learning menawarkan keuntungan signifikan dibandingkan pemrograman tradisional untuk masalah tertentu:

*   **Menyederhanakan Sistem Berbasis Aturan yang Kompleks**: Untuk masalah seperti penyaringan spam, metode tradisional melibatkan daftar aturan yang panjang dan sulit dipelihara. ML secara otomatis mempelajari pola, menghasilkan program yang lebih sederhana, lebih akurat, dan lebih mudah dipelihara.
*   **Beradaptasi dengan Perubahan**: Sistem ML dapat secara otomatis beradaptasi dengan pola baru (misalnya, *spammer* mengubah taktik) tanpa intervensi manual, tidak seperti sistem berbasis aturan yang membutuhkan pembaruan konstan.
*   **Memecahkan Masalah Tanpa Algoritma yang Diketahui**: Untuk masalah kompleks seperti pengenalan ucapan, di mana aturan *hardcoding* tidak mungkin dilakukan, ML dapat belajar dari data untuk menemukan solusi.
*   **Mendapatkan Wawasan (Data Mining)**: Dengan memeriksa model ML yang terlatih, manusia dapat menemukan korelasi yang tidak terduga, tren baru, dan mendapatkan pemahaman yang lebih baik tentang masalah dan data.

## Contoh Aplikasi

Machine Learning diterapkan dalam berbagai tugas:

*   **Klasifikasi Gambar**: Menganalisis gambar produk di lini produksi untuk mengklasifikasikannya secara otomatis (misalnya, menggunakan Convolutional Neural Networks - CNNs).
*   **Segmentasi Semantik**: Mendeteksi tumor dalam pemindaian otak dengan mengklasifikasikan setiap piksel (misalnya, menggunakan CNNs).
*   **Klasifikasi Teks**: Mengklasifikasikan artikel berita secara otomatis atau menandai komentar yang menyinggung (misalnya, menggunakan Recurrent Neural Networks - RNNs, CNNs, atau Transformers).
*   **Rangkuman Teks**: Meringkas dokumen panjang secara otomatis (menggunakan alat NLP seperti RNNs, CNNs, atau Transformers).
*   **Chatbot/Asisten Pribadi**: Melibatkan komponen NLP seperti Natural Language Understanding (NLU) dan modul tanya jawab.
*   **Peramalan**: Memprediksi pendapatan perusahaan tahun depan, berdasarkan banyak metrik kinerja (menggunakan model Regresi seperti Regresi Linier, SVM, Random Forest, atau Jaringan Saraf Tiruan).
*   **Pengenalan Ucapan**: Membuat aplikasi bereaksi terhadap perintah suara (membutuhkan pemrosesan sampel audio; biasanya diproses menggunakan RNNs, CNNs, atau Transformers).
*   **Deteksi Anomali**: Mendeteksi penipuan kartu kredit atau cacat manufaktur.
*   **Clustering**: Mengelompokkan klien berdasarkan pembelian untuk strategi pemasaran yang ditargetkan.
*   **Visualisasi Data/Reduksi Dimensi**: Merepresentasikan dataset kompleks berdimensi tinggi secara jelas dan informatif (misalnya, menggunakan PCA, t-SNE).
*   **Sistem Rekomendasi**: Menyarankan produk yang mungkin diminati klien, berdasarkan pembelian sebelumnya (misalnya, menggunakan Artificial Neural Networks).
*   **Membangun Bot Cerdas untuk Game**: Seringkali ditangani menggunakan Reinforcement Learning (RL), di mana agen belajar strategi untuk memaksimalkan *reward* seiring waktu (misalnya, AlphaGo).

## Jenis Sistem Machine Learning

Sistem ML dapat dikategorikan berdasarkan kriteria berikut:

*   Apakah mereka dilatih dengan pengawasan manusia (Supervised, Unsupervised, Semisupervised, dan Reinforcement Learning).
*   Apakah mereka dapat belajar secara bertahap saat itu juga (Online versus Batch Learning).
*   Apakah mereka bekerja dengan membandingkan titik data baru dengan titik data yang diketahui, atau dengan mendeteksi pola dalam data pelatihan dan membangun model prediktif (Instance-Based versus Model-Based Learning).

Kriteria ini tidak eksklusif; Anda dapat menggabungkannya sesuka Anda. Contohnya, filter spam canggih mungkin belajar dengan cepat menggunakan model jaringan saraf dalam yang dilatih menggunakan contoh spam dan non-spam; ini menjadikannya sistem pembelajaran *online*, berbasis model, dan *supervised*.

### Supervised Learning

Dalam *supervised learning*, set pelatihan yang Anda berikan kepada algoritma mencakup solusi yang diinginkan, yang disebut **label**.

*   **Klasifikasi**: Memprediksi label kategorikal (misalnya, spam/bukan spam, kategori gambar).
    *   **Contoh Algoritma**: k-Nearest Neighbors, Regresi Linier (dapat digunakan untuk klasifikasi), Regresi Logistik, Support Vector Machines (SVMs), Decision Trees, Random Forests, Neural Networks.
*   **Regresi**: Memprediksi nilai numerik target (misalnya, harga rumah, harga mobil).
    *   **Contoh Algoritma**: Regresi Linier, Regresi Polinomial, Regresi SVM, Regresi Random Forest, Neural Networks.

### Unsupervised Learning

Dalam *unsupervised learning*, data pelatihan **tidak berlabel**, dan sistem mencoba mempelajari pola dan struktur dengan sendirinya, tanpa "guru."

*   **Clustering**: Mengelompokkan instansi serupa.
    *   **Contoh Algoritma**: K-Means, DBSCAN, Hierarchical Cluster Analysis (HCA).
*   **Deteksi Anomali & Deteksi Kebaruan**: Mengidentifikasi instansi yang tidak biasa yang menyimpang secara signifikan dari norma.
    *   **Contoh Algoritma**: One-class SVM, Isolation Forest.
*   **Visualisasi dan Reduksi Dimensi**: Mengurangi jumlah fitur sambil mempertahankan informasi penting, seringkali untuk visualisasi yang lebih mudah.
    *   **Contoh Algoritma**: Principal Component Analysis (PCA), Kernel PCA, Locally Linear Embedding (LLE), t-Distributed Stochastic Neighbor Embedding (t-SNE).
    *   **Ekstraksi Fitur**: Menggabungkan fitur-fitur yang berkorelasi menjadi satu (jenis reduksi dimensi).
*   **Pembelajaran Aturan Asosiasi**: Menemukan hubungan menarik antara atribut dalam dataset besar (misalnya, "orang yang membeli X juga cenderung membeli Y").
    *   **Contoh Algoritma**: Apriori, Eclat.

### Semisupervised Learning

Pendekatan ini menangani data yang **sebagian berlabel**, biasanya sejumlah besar data tidak berlabel dan sejumlah kecil data berlabel. Sebagian besar algoritma *semisupervised learning* adalah kombinasi dari teknik *unsupervised* dan *supervised*.

*   **Contoh**: Google Photos menggunakan *clustering* (unsupervised) untuk mengelompokkan foto orang yang sama, lalu Anda memberikan beberapa label (supervised) untuk memberi nama semua orang di semua foto.
*   **Algoritma**: Deep Belief Networks (DBNs) berdasarkan Restricted Boltzmann Machines (RBMs).

### Reinforcement Learning

Paradigma yang sangat berbeda di mana sistem pembelajaran, yang disebut **agen** dalam konteks ini, dapat mengamati **lingkungan**, memilih dan melakukan **tindakan**, dan mendapatkan **hadiah** sebagai imbalan (atau hukuman dalam bentuk hadiah negatif). Agen tersebut kemudian harus belajar sendiri apa **strategi (policy)** terbaik untuk mendapatkan *reward* terbanyak seiring waktu. Sebuah *policy* mendefinisikan tindakan apa yang harus dipilih agen ketika berada dalam situasi tertentu.

*   **Contoh**: Robot belajar berjalan, AI game (misalnya, AlphaGo mengalahkan juara dunia Go).

## Batch dan Online Learning

Klasifikasi ini bergantung pada apakah sistem dapat belajar secara bertahap dari aliran data yang masuk.

### Batch Learning (Offline Learning)

Sistem dilatih menggunakan **semua data yang tersedia sekaligus**, biasanya *offline*. Tidak dapat belajar secara bertahap.

*   **Proses**: Latih pada seluruh dataset -> Luncurkan ke produksi -> Sistem berjalan tanpa pembelajaran lebih lanjut.
*   **Adaptasi terhadap Data Baru**: Memerlukan pelatihan ulang sistem dari awal pada seluruh dataset (yang diperbarui), lalu mengganti sistem lama dengan yang baru.
*   **Keuntungan**: Lebih sederhana untuk diimplementasikan.
